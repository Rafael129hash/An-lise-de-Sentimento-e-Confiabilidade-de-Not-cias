```python
import pandas as pd
import re

# Carregar os dados (substitua 'caminho_do_arquivo.csv' pelo caminho correto do seu arquivo)
df = pd.read_csv("fake_news_dataset.csv")

# Função para remover caracteres indesejados (pontuação, etc.)
def clean_text(text):
    # Remover caracteres não alfanuméricos e substituir por espaço
    return re.sub(r'[^a-zA-Z0-9\s]', '', text)

# Aplicar a função de limpeza nas colunas de interesse
df['title'] = df['title'].apply(lambda x: clean_text(x))
df['author'] = df['author'].apply(lambda x: clean_text(x))
df['state'] = df['state'].apply(lambda x: clean_text(x))
df['category'] = df['category'].apply(lambda x: clean_text(x))

# Verificar as primeiras linhas após limpeza
print(df[['title', 'author', 'state', 'category']].head())

# Verificar valores nulos
print("\nValores nulos por coluna:")
print(df.isnull().sum())

# Verificar duplicatas
print("\nNúmero de duplicatas:", df.duplicated().sum())

# Verificar tipos de dados incorretos
print("\nTipos de dados da tabela:")
print(df.dtypes)
```

```python
# Salvar a tabela ajustada em um arquivo CSV
df.to_csv('tabela_ajustada.csv', index=False)  # O argumento index=False evita salvar o índice como uma coluna no arquivo
```
